```python
 # CrÃ©ation de la fonction permettant de tracer un vecteur vitesse
def vecteur_vitesse(x, y, dt, i) :
    vx = (x[i+1] - x[i]) / dt 
    vy = (y[i+1] - y[i]) / dt   

    quiver(x[i], y[i], vx, vy, scale_units='xy',angles='xy',color='blue',width=0.010,scale=4)    
    vitesse = sqrt(((x[i+1]-x[i])/(dt))**2 +((y[i+1]-y[i])/(dt))**2) # Calcul de la vitesse au point i
    print("A la position M",i, " la vitesse est de",round(vitesse,1), "m/s")
    text(x[i]+0.20,y[i]+0.50,r"$\vec{v}$"+str(i),color="blue") #permet d'afficher le nom Ã  cÃ´tÃ© du vecteur
    
#Affichage 
plot(x, y, 'r--o')
vecteur_vitesse(x, y, dt, 0) #tracÃ© du vecteur vitesse au point M0
vecteur_vitesse(x, y, dt, 1) #tracÃ© du vecteur vitesse au point M1
vecteur_vitesse(x, y, dt,  ) #tracÃ© du vecteur vitesse au point M2
vecteur_vitesse(x, y, dt,  ) #tracÃ© du vecteur vitesse au point M3
vecteur_vitesse(x, y, dt,  ) #tracÃ© du vecteur vitesse au point M4

ylim(0,6)
grid(True)
xlabel(" Position x (m) ")
ylabel(" Position y (m) ")
title(" ******** ")
show()
```
